# Overall Engagement Rate
In order to comprehend how various customers behave and respond to various marketing strategies, we will analyze it.

```python
#Information about whether a customer responded to the marketing efforts can be found in the Response field.
#Find out how many customers have responded in total.
df.groupby('Response').count()['Customer']
```

```python
#Use a bar plot to visualize the same thing.
ax = df.groupby('Response').count()['Customer'].plot(kind='bar', color = 'red',grid=True, inline  figsize=(10,6),
title='Marketing Engagement')

ax.set_xlabel('Engaged')
ax.set_ylabel('Count')

plt.show()
```

```python
#Determine the proportions of engaged and unengaged customers.
df.groupby('Response').count()['Customer']/df.shape[0]
```

We can see from this output and the plot that only about 14% of customers answered the marketing calls.

# Engagement Rates by Offer Type
This DataFrame's Renew Offer Type column details the kind of renewal offer offered to customers. We will investigate the kinds of offers that were most effective for the engaged customers.

```python
df.groupby('Renew Offer Type').count()['Customer']
```

```python
#Finding the engagement rates for each type of renewal offer.
by_offer_type_df = df.loc[
    df['Response'] == 'Yes' # Counting only agreed/engaged customers
].groupby([ 
    'Renew Offer Type' # engaged customers grouped by renewal offer type
]).count()['Customer']/df.groupby('Renew Offer Type').count()['Customer']


by_offer_type_df
```

```python
# Visualize this distribution
ax = (by_offer_type_df*100.0).plot(kind = 'bar',figsize = (10,6),grid = True,color = 'palegreen',)

ax.set_ylabel('Engagement Rate (%)')

plt.show()
```

This demonstrates that Offer 2 is superior.

# Offer Type & Vehicle Class
We will learn how different types of customers respond to different marketing messages in different ways. We begin taking a gander at the commitment rates by each proposition type and vehicle class

```python
by_offer_type_df = df.loc[
    df['Response'] == 'Yes' # engaged Customers
].groupby(['Renew Offer Type', 'Vehicle Class' ## grouping the data by these two columns
          ]).count()['Customer']/df.groupby(['Renew Offer Type']).count()['Customer']

by_offer_type_df

# Note: Divided by df.groupby(['Renew Offer Type']).count()['Customer'] --> Is for percentage calculation
```

```python
#Utilize the unstack function to make the previous output easier to read.
#to extract and transform the inner-level groups into columns, pivot the data
by_offer_type_df = by_offer_type_df.unstack().fillna(0)
by_offer_type_df
```

```python
#Create a bar plot to represent this data.
ax = (by_offer_type_df*100.0).plot(kind='bar',figsize=(10, 8),grid=True)
ax.set_ylabel('Engagement Rate (%)')
plt.show()
```

The previous section, "Engagement Rates by Offer Type," revealed that Offer2 had the highest customer response rate. With the breakdown of the customer attributes into the category "Vehicle class," we are now able to add more insights:

We can see that customers with a "Luxury SUV" respond more frequently to Offer1 than to Offer2, and that customers with a "Four-Door Car" respond more frequently to all offer types. We can fine-tune who to target with various offers if the response rates of various customer rates differ significantly.

# Engagement Rates by Sales Channel

```python
#We'll look at how different sales channels affect engagement rates.
by_sales_channel_df = df.loc[
df['Response'] == 'Yes'
].groupby([
'Sales Channel'
]).count()['Customer']/df.groupby('Sales Channel').count()['Customer']

by_sales_channel_df
```

```python
#Use a bar chart to represent this.
ax = (by_sales_channel_df*100.0).plot(
kind='bar',
figsize=(10, 8),
color='plum',
grid=True
)
ax.set_ylabel('Engagement Rate (%)')
plt.show()
```

As can be seen, when it comes to getting responses from customers, Agent performs better than Web-based sales.

# Sales Channel & Vehicle Size
We'll find out if customers with a variety of vehicle sizes react differently to various sales channels.

```python
by_sales_channel_df = df.loc[
df['Response'] == 'Yes'
].groupby([
'Sales Channel', 'Vehicle Size'
]).count()['Customer'] / df.groupby('Sales Channel').count()['Customer']

by_sales_channel_df
```

```python
# Unstack the data into a more visible format
by_sales_channel_df = by_sales_channel_df.unstack().fillna(0)
by_sales_channel_df
```

```python
#Use a bar chart to represent this.
ax = (by_sales_channel_df*100.0).plot(
kind='bar',
figsize=(10, 8),
grid=True,)

ax.set_ylabel('Engagement Rate (%)')
plt.show()
```

# Engagement Rates by Months Since Policy Inception

```python
by_months_since_inception_df = df.loc[
df['Response'] == 'Yes'
].groupby(by='Months Since Policy Inception'
)['Response'].count() / df.groupby(
by='Months Since Policy Inception'
)['Response'].count() * 100.0


by_months_since_inception_df.fillna(0)
```

```python
# Visualize this using a plot
ax = by_months_since_inception_df.fillna(0).plot(
figsize=(15, 10),
title='Engagement Rates by Months Since Inception',
grid=True,
color='slateblue'
)
ax.set_xlabel('Months Since Policy Inception')
ax.set_ylabel('Engagement Rate (%)')
plt.show()
```

# Customer Segmentation by CLV & Months Since Policy Inception

```python
## Take a look at the distribution of the CLV
df['Customer Lifetime Value'].describe()
```

We will classify customers whose CLV is higher than the median as high-CLV customers and lower than the median as low-CLV customers for the previous output.

```python
df['CLV Segment'] = df['Customer Lifetime Value'].apply(
lambda x: 'High' if x > df['Customer Lifetime Value'].median() else 'Low') 
```

```python
# Do the same procedure for Months Since Policy Inception
df['Months Since Policy Inception'].describe()
```

```python
df['Policy Age Segment'] = df['Months Since Policy Inception'].apply(
lambda x: 'High' if x > df['Months Since Policy Inception'].median() else 'Low')
```

```python
df.head()
```

Visualize segmentation

```python
ax = df.loc[
(df['CLV Segment'] == 'High') & (df['Policy Age Segment'] == 'High')
].plot.scatter(
x='Months Since Policy Inception',
y='Customer Lifetime Value',
logy=True,
color='red'
)
df.loc[
(df['CLV Segment'] == 'Low') & (df['Policy Age Segment'] == 'High')
].plot.scatter(
ax=ax,
x='Months Since Policy Inception',
y='Customer Lifetime Value',
logy=True,
color='blue'
)
df.loc[
(df['CLV Segment'] == 'High') & (df['Policy Age Segment'] == 'Low')
].plot.scatter(
ax=ax,
x='Months Since Policy Inception',
y='Customer Lifetime Value',
logy=True,
color='orange'
)
df.loc[
(df['CLV Segment'] == 'Low') & (df['Policy Age Segment'] == 'Low')
].plot.scatter(
ax=ax,
x='Months Since Policy Inception',
y='Customer Lifetime Value',
logy=True,
color='green',
grid=True,
figsize=(10, 7)
)

ax.set_ylabel('CLV (in log scale)')
ax.set_xlabel('Months Since Policy Inception')
ax.set_title('Segments by CLV and Policy Age')

plt.show()
```
